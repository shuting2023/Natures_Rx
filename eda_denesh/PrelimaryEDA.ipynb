```python
plt.figure(figsize=(10, 6))
plt.scatter(usdf['GCPNT_LON'], usdf['GCPNT_LAT'], alpha=0.5)
plt.title('Geographical Distribution of Urban Centers')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
corr_matrix = rawdf[greencols].corr()
corr_matrix
```
